```python
df=df[df['native-country']!='?']
df=df[df['occupation']!='?']
```

```python
cat_data=df.select_dtypes('object')
cat_data.apply(lambda x:x=='?').sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cat_data_le=cat_data.apply(le.fit_transform)
```

```python
df=df.drop(cat_data_le.columns,axis=1)
df=pd.concat([df,cat_data_le],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('income',axis=1)
Y=df['income']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=5)
dt.fit(X_train,Y_train)
```

```python
;from sklearn import metrics
Y_pred=dt.predict(X_test)
print(metrics.classification_report(Y_test,Y_pred))
```

```python

print(metrics.accuracy_score(Y_test,Y_pred))
dt.score(X_test,Y_test)
```

```python
print(metrics.confusion_matrix(Y_test,Y_pred))
```

```python

from sklearn.model_selection import GridSearchCV
params={'max_depth':range(1,41)}
n_folds=5
dtree=DecisionTreeClassifier(random_state=69)
cv_tree=GridSearchCV(dtree,param_grid=params,scoring='accuracy',return_train_score=True,verbose=1,cv=n_folds,n_jobs=-1)
cv_tree.fit(X_train,Y_train)
```

```python
result=pd.DataFrame(cv_tree.cv_results_)
result
```

```python
sns.lineplot(x=result['param_max_depth'],y=result['mean_train_score'])
sns.lineplot(x=result['param_max_depth'],y=result['mean_test_score'])
plt.grid()
```

```python
cv_tree.best_params_
```

```python
cv_tree.best_score_
```

```python
# tuning min_samples_leaf
params={'min_samples_leaf':range(5,200,20)}
n_folds=5
dt=DecisionTreeClassifier(random_state=69)
model=GridSearchCV(dt,param_grid=params,scoring='accuracy',verbose=1,return_train_score=True,cv=n_folds)
model.fit(X_train,Y_train)
```

```python
result=pd.DataFrame(model.cv_results_)
result
```

```python
sns.lineplot(x=result['param_min_samples_leaf'],y=result['mean_train_score'])
sns.lineplot(x=result['param_min_samples_leaf'],y=result['mean_test_score'])
plt.grid()
```

```python
# min samples leaf is 25
params={'min_samples_split':range(5,300,20)}
dt=DecisionTreeClassifier(random_state=69)
model_cv=GridSearchCV(dt,param_grid=params,scoring='accuracy',return_train_score=True,cv=n_folds,verbose=1)
model_cv.fit(X_train,Y_train)
```

```python
result=pd.DataFrame(model_cv.cv_results_)
result
```

```python
sns.lineplot(x=result['param_min_samples_split'],y=result['mean_train_score'])
sns.lineplot(x=result['param_min_samples_split'],y=result['mean_test_score'])
plt.grid()
```
